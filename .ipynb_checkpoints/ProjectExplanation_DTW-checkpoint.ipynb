# Project Overview

Use Dynamic Time Warping to compare n sized stock signals to historical signals of the same size to try and predict
if the next day will end up positive or negative. The closer the signals match a historical signal, the lower the
cost to dynamicly time warp the signal.

# Dynamic Time Warping Background

Dyanimic Time Warping (DTW) is an algorithm used to measure similarity between two signals. These signals are warped non-linearly to determine a measure of their similarity independent of certain non-linear variations in the time dimension.

Signal is converted to percent change from first point.
<br> Orange is the signal we are trying to predict the next day for while blue is the historical signal we are comparing the prediction signal to.

Dynamic time warping will get cost to move a single point to all other points to compare signal. 
<br> Each cell relates to distance between those points in the two signals.

<div>

</div>

Once matrix is created the path of least cost between point (0,0) and (n,m) will be calculated. 
<br> In this case it is least cost path between (0,0) and (20,20) because both signals have a length of 20 points.
<br> Signal will be moved n points over and comparison will be repeated until there are no more signals in historical data.

# Code Step-by-step

1. Pull all historical data currently available for a certain ticker and for S&P
2. Divide historical data into signals we want to predict next day for vs signals we want to compare the prediction to
3. Take one prediction signal and collect average cost to dynamically time warp the prediction signal to compare signal, next day percent up after copmare signal, and actual next day percent up if cost is below certain threshold
4. Move compare signal n data points over and compare and collect data again
5. Evaluate predicted data based on:
    - if average next day from historical is positive
    - if average next day s&p from historical is positive
6. Evaluate profitability

# Input Parameter Analysis

nPoints - signal size
<br> Analyzed nPoints = 20 - 100 by steps of 10 to find maximum returns on ticker 'APPL' with a cost compare value of 0.5
<br> Will need to analyze other tickers to see if it is a good n value for all but for now it looks like 50 is a good value for n. Larger values take considerably longer.

CostCompareValue - if path cost for matching is less than this value than we collect for our averages

<br> This is for 'APPL' ticker - will need to see if this spans across all stocks
